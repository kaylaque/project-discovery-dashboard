```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
# url = 'https://www.imdb.com/list/ls024149810/'
url = 'https://old.igem.org/Team_List?year=2023'
# url = 'https://reactjs.org'
r = requests.get(url)

```

```python
print(r.status_code)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
all = re.split(r"<tr>", str(soup.table.tbody.tr))[1:]
len(all)
```

```python

def extract_each_row(row):
    # Create a BeautifulSoup object
    soup = BeautifulSoup(row, 'html.parser')

    # Extract the team name
    team_name = soup.find('a', class_='team_name').get_text()
    link_team = soup.find('a', class_='team_name').get('href')
    wiki_url = soup.find('a', href=re.compile('wiki')).get('href')

    # extract other information
    infos = []
    for a in row.split('<td data-order='):
        infos.append(a.split('>')[-1])
    region = infos[2]
    country = infos[3]
    track = infos[4]
    kind = infos[5]
    section = infos[6]
    application = infos[7]
    return(team_name, link_team, wiki_url, region, country, track, kind, section, application)
```

```python
teams = []
links = []
wikis = []
regions = []
countries = []
tracks = []
kinds = []
sections = []
apps = []
for i in range(len(all)):
    team_name, link_team, wiki_url, region, country, track, kind, section, application = extract_each_row(all[i])
    teams.append(team_name)
    links.append(link_team)
    wikis.append(wiki_url)
    regions.append(region)
    countries.append(country)
    tracks.append(track)
    kinds.append(kind)
    sections.append(section)
    apps.append(application)
```

```python
# Create a DataFrame
df = pd.DataFrame({
    'Team name': teams,
    'Profile team': links,
    'Wiki URL': wikis,
    'Region': regions,
    'Country': countries,
    'Track': tracks,
    'Kind': kinds,
    'Section': sections,
    'Application': apps
})

# Print the DataFrame
df
```

```python
df.to_csv('teams_col_basic.csv', index=False)
```
